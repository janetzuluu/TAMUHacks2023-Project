```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s') #define model
```

```python
img = 'https://classicalatelierathome.com/wp-content/uploads/2013/03/squinteyes.jpg' #images for analysis
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
# testing detection of persons and items
```

```python
results.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, distance = cap.read()
    
    #Detect motion
    results = model(distance)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images')
labels = ['awake', 'sleepy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        ret, frame = cap.read()
        
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
